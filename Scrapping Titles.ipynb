# Scrap Title One Piece Episode
Scrapping all title one piece episode

## Initialize Library

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

## Get Titles

```python
page = requests.get("https://faisalanimedroidblogspot.wordpress.com/2018/02/15/kumpulan-judul-episode-one-piece-lengkap-update-2018/")
html_page = page.content
soup = BeautifulSoup(html_page, "lxml")
titles1 = soup.find_all('h5')

page = requests.get("https://faisalanimedroidblogspot.wordpress.com/2018/02/15/kumpulan-judul-episode-one-piece-lengkap-update-2018-part-2/")
html_page = page.content
soup = BeautifulSoup(html_page, "lxml")
titles2 = soup.find_all('h5')

titles = titles1 + titles2
#titles
```

## Fix Structure

```python
#type(titles)
pd.set_option('display.max_colwidth', -1) # displayed full

df = pd.DataFrame(titles)
df = df[0] # set first column
```

```python
# check data

# for count, item in enumerate(df, start=0):
#     print(count, item)
```

```python
def split_data(df):
    data = str(df)
    dataSplit = data.split("<br/>")
    dataList = [x.replace("\n","") for x in dataSplit]
    
    df = pd.DataFrame(dataList)
    
    return df
```

```python
# problem in index: 230, 427, (delete 453-463), 527, 608
data1 = df[0:230]
data2 = split_data(df[230])
data3 = df[231:427]
data4 = split_data(df[427])
data5 = df[428:453]
data6 = df[464:527]
data7 = split_data(df[527])
data8 = df[528:608]
data9 = split_data(df[608])
data10 = df[609:692]

# combining data
dataAll = data1.append(data2, ignore_index=True)
dataAll = dataAll[0]

dataAll = dataAll.append(data3, ignore_index=True)
dataAll = dataAll.append(data4, ignore_index=True)
dataAll = dataAll[0]

dataAll = dataAll.append(data5, ignore_index=True)
dataAll = dataAll.append(data6, ignore_index=True)
dataAll = dataAll.append(data7, ignore_index=True)
dataAll = dataAll[0]

dataAll = dataAll.append(data8, ignore_index=True)
dataAll = dataAll.append(data9, ignore_index=True)
dataAll = dataAll[0]

dataAll = dataAll.append(data10, ignore_index=True)

#print(dataAll)
```

```python
# for count, item in enumerate(dataAll, start=0):
#     print(count, item)
```

```python
# fix Episode 781
dataAll.loc[784] = '<h5>Episode 781 ” Tiga Orang yang Keras Kepala! Pengejaran Besar Mengejar Kelompok Topi Jerami! “</h5>'

#Drop data
dataAll = dataAll.drop([3], axis=0)
dataAll = dataAll.drop([9], axis=0)
dataAll = dataAll.drop([47], axis=0)
dataAll = dataAll.drop([209], axis=0)
dataAll = dataAll.drop([455], axis=0)

# remove first 542: Reuni Toriko
dataAll = dataAll.drop([545], axis=0)

dataAll = dataAll.drop([785], axis=0)
dataAll = dataAll.drop([786], axis=0)
dataAll = dataAll.drop([797], axis=0)

# Change 662 -> 362
dataAll.loc[365] = '<h5>Episode 362: ” Tarian Pedang Diatas Atap! Akhir Pertarungan Zoro Melawan Ryuuma!”</h5>'

dataAll = dataAll.reset_index(drop=True)
#print(dataAll)
```

```python
# for count, item in enumerate(dataAll, start=0):
#     print(count, item)
```

```python
dataTitles = dataAll.copy()

# initial data Episode 452, 571 
d452 = ['<h5>Episode 452: ”Menuju Markas Besar Angkatan Laut! Perjalanan Menyelamatkan Ace! “</h5>']
data452 = pd.DataFrame(d452)

d571 = ['<h5>Episode 571: ”Dia Menyukai Manisan! Big Mam Salah Satu dari Yonkou! “</h5>']
data571 = pd.DataFrame(d571)

dataTitles1 = dataTitles[0:451]
dataTitles2 = dataTitles[451:569]
dataTitles3 = dataTitles[569:824]

# add data
dataFix = dataTitles1.append(data452, ignore_index=True)
dataFix = dataFix[0]

dataFix = dataFix.append(dataTitles2, ignore_index=True)
dataFix = dataFix.append(data571, ignore_index=True)
dataFix = dataFix[0]

dataFix = dataFix.append(dataTitles3, ignore_index=True)

#dataFix
```

## Cleaning

```python
df = dataFix.copy()

df = df.apply(lambda x:str(x).strip()) #remove leading and trailing whitespaces
df = df.apply(lambda x:str(x).replace("<h5>","").replace("</h5>",""))
df = df.apply(lambda x:str(x).replace("“","").replace("”","")) # remove Quotation marks
df = df.apply(lambda x:str(x).replace("\"","")) # remove Quotation marks
df = df.apply(lambda x:str(x).replace("<br/>","").replace("<b>","").replace("</b>","")) # remove tag br and b
df = df.apply(lambda x:str(x).replace("<h5 style=\"text-align:left;\">","")) # remove tag
df = df.apply(lambda x:str(x).replace("  "," ")) # remove double space

df = df.apply(lambda x:str(x).replace(":"," ",1)) # replace : with space
df = df.apply(lambda x:str(x).replace("   "," ")) # remove triple space
df = df.apply(lambda x:str(x).replace("  "," ")) # remove double space
#df = df.apply(lambda x:str(x).replace(":"," ",1))

df = df.apply(lambda x:str(x).replace("Episdoe","Episode").replace("Episoe","Episode").replace("DresrossaEpisode","Episode")) # fix the 'Episode'
df = df.apply(lambda x:str(x).replace("Episode ","", 1)) # replace first occurence of 'Episode'
df = df.apply(lambda x:str(x).replace(" lanan Menyelamatkan Ace","")) # remove error Title


#print(df[0:50])
```
